```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def initialisation_Kcentroids(A, K):
    l = len(A)
    return A[np.random.choice(l, K, replace=False), :]

def mean(A, i, K):
    m, n = A.shape
    centroids = np.zeros((K, n))
    for k in range(K):
        examples = A[np.where(i == k)]
        mean = [np.mean(column) for column in examples.T]
        centroids[k] = mean
    return centroids

img=Image.open('/content/43.jpg')
img=np.asarray(img) /255
w, h, d = img.shape
A = img.reshape((w * h, d))
K = 15                         # the desired number of colors(cluster) in the compressed image
centroids = initialisation_Kcentroids(A, K)
previous_centroids = centroids
max_iteration=10
for _ in range(max_iteration):
  l = len(A)
  c = np.zeros(l)
  for i in range(l):
    distances = np.linalg.norm(A[i] - centroids, axis=1)
    # Assign closest cluster to c[i]
    c[i] = np.argmin(distances)
  ix = c
  centroids = mean(A, ix, K)
  if (centroids == previous_centroids).all():
    # The centroids aren't moving anymore.
    colors=centroids
  else:
    previous_centroids = centroids
colors, m =centroids, ix
ix = np.array(ix, dtype=np.uint8)
A_reconstructed = np.array(colors[ix, :] * 255, dtype=np.uint8).reshape((w, h, d))
compressed_image = Image.fromarray(A_reconstructed)
plt.title("k=15")
plt.imshow(compressed_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('/content/43.jpg')    #reading the image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)     #color change from BGR to RGB
pixv = img.reshape((-1,3))       # Reshaping the image into a 2D array of pixels and 3 color values (RGB)
pixv = np.float32(pixv)     # Changing data type to float type
cond = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85) # here the code is the condition for the algorithm to stop running

k = 10 # then perform k-means clustering with number of clusters defined as 5 here
retval, labels, centers = cv2.kmeans(pixv, k, None, cond, 10, cv2.KMEANS_RANDOM_CENTERS)  #kmeans clustring
centers = np.uint8(centers)  #data converted into 8-bit values
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape((img.shape)) # reshape data into the original image dimensions

plt.imshow(segmented_image)
plt.title("k=20")
plt.savefig("k5.png")
```
